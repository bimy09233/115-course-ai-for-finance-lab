```python
import ipywidgets as widgets
```

### 本息平均攤還

$$c=P_0(\frac{r}{1-(1+r)^{-N}})$$

$P_0=貸款金額$

$r=月利率$

$N=期數$

```python
P = widgets.IntText(description="貸款總金額")
Y = widgets.IntText(description="貸款年限(年)")
G = widgets.IntText(description="寬限期(年)")
R = widgets.FloatText(description="年利率(%)")
calculate_button = widgets.Button(
    description="計算",
    button_style="primary",
    icon="calculator",
)

result_output = widgets.Output()

def year_to_month_rate(R):
    return R.value / 100 / 12

def payment_periods(_N):
    return _N.value * 12

def mortgage_calculator(_button):
    _P0 = float(P.value)
    _r = float(year_to_month_rate(R))
    _g = float(payment_periods(G))
    _y = float(payment_periods(Y)) - _g
    
    monthly_payment_interest = round(_P0 * _r)
    monthly_payment = round((_r / (1 - (1 + _r)**(-_y))) * _P0)
    
    with result_output:
        result_output.clear_output()

        if _g > 0:
            display(f"寬限期月繳：{monthly_payment_interest}")
        display(f"本息攤月繳：{monthly_payment}")
        display("-----------------------")
        
        for i in range(1, int(_g + _y) + 1, 1):
            if i <= _g:
                display(f"第{("0"+str(i))[-2:]}期：{monthly_payment_interest}")
            else:
                display(f"第{("0"+str(i))[-2:]}期：{monthly_payment}")
    
calculate_button.on_click(mortgage_calculator)
display(widgets.VBox([P, Y, G, widgets.HBox([R, calculate_button]), result_output]))
```
