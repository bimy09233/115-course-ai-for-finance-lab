### 不同頻率複利計息

```python
P = 1_000_000 # 本金
R = 0.1       # 年息
n = [1, 2, 4, 12, 52, 365]
freq = ['年', '半年', '季', '月', '週', '日']
```

```python
for i, j in zip(n, freq):
    interest_rate = R / i
    interest = P * interest_rate
    total_interest = P * (1 + interest_rate)**i - P
    
    print(f'{j}利率：{interest_rate}')
    print(f'每{j}利息：{interest:,.0f}')
    print(f'總利息：{total_interest:,.0f}')
    print('-----------------------')
```
